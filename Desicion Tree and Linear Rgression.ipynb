```python
def entropy(elements):
    counter = Counter(elements)
    probs = [counter[c] / len(elements) for c in set(elements)]
#     ic(probs)
    return - sum(p * np.log(p) for p in probs)
```

```python
def find_the_optimal_spliter(training_data: pd.DataFrame, target: str):
    if not training_data.empty:
        x_fileds = set(training_data.columns.tolist()) - {target}
        spliter = None
        min_entropy = float('inf')
        for f in x_fileds:
    #         ic(f)
            values = set(training_data[f])
    #         ic(values)
            for v in values:

                sub_spliter_1 = training_data[training_data[f] == v][target].tolist()
    #             ic(sub_spliter_1)
                entropy_1 = entropy(sub_spliter_1)
    #             ic(entropy_1)

                sub_spliter_2 = training_data[training_data[f] != v][target].tolist()
    #             ic(sub_spliter_2)
                entropy_2 = entropy(sub_spliter_2)
    #             ic(entropy_2)

                entropy_v = entropy_1 + entropy_2
    #             ic(entropy_v)

                if entropy_v <= min_entropy:
                    min_entropy = entropy_v
                    spliter = (f, v)

        print('spliter is: {}'.format(spliter))
        print('the min entropy is: {}'.format(min_entropy))
#     else:
#         return
    return spliter
```

```python
def find_all_optimal_spliter(training_data: pd.DataFrame, target: str):
    spliter0 = find_the_optimal_spliter(training_data, target='bought')
#     print(spliter0)
    dataset1 = training_data[training_data[spliter0[0]] == spliter0[1]]
    print(dataset1)
    print(len(dataset1))
    dataset2 = training_data[training_data[spliter0[0]] != spliter0[1]]
    print("===================")
    print("===================")
    print(dataset2)
    print(len(dataset2))
#     print(len(set(dataset1['bought'])))
#     print(len(set(dataset2['bought'])))
    if len(set(dataset1['bought'])) != 1:
        if not dataset1.empty:
            spliter1 = find_all_optimal_spliter(dataset1, target='bought')
            print('spliter1 is: {}'.format(spliter1))
    if len(set(dataset2['bought'])) != 1:
        if not dataset2.empty:
            spliter2 = find_all_optimal_spliter(dataset2, target='bought')
            print('spliter2 is: {}'.format(spliter2))


#     print('spliter0 is: {}'.format(spliter0))
#     print('spliter1 is: {}'.format(spliter1))
#     print('spliter2 is: {}'.format(spliter2))
#     print('spliter11 is: {}'.format(spliter11))
#     print('spliter12 is: {}'.format(spliter12))
#     print('spliter21 is: {}'.format(spliter21))
#     print('spliter22 is: {}'.format(spliter22))
#     return 
```

## Linear Regression

```python
import numpy as np
```

```python
random_data = np.random.random((30,2))
```

```python
import matplotlib.pyplot as plt
```

```python
X = random_data[:, 0]
```

```python
y = random_data[:, 1]
```

```python
import random
```

```python
def assmuing_function(x):
    return 13.4 * x + 5 + random.randint(-5, 5)
```

```python
y = [assmuing_function(x) for x in X]
```

```python
plt.scatter(X, y)
```

```python
y = np.array(y)
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg =  LinearRegression().fit(X.reshape(-1, 1), y)
```

```python
reg.score(X.reshape(-1, 1), y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
def f(x):
    return reg.coef_ * x + reg.intercept_
```

```python
def loss(y,y_hat):
    return sum(abs(y_i - y_hat_i) for y_i, y_hat_i in zip(list(y),list(y_hat)))/len(list(y))
```

```python
def partial_derivative_k(x, y, y_hat): 
    n = len(y)
    gradient = 0
    for x_i, y_i, y_hat_i in zip(list(x),list(y),list(y_hat)):
        gradient += np.sign(y_i-y_hat_i) * x_i
    return 1/n * gradient

def partial_derivative_b(y, y_hat):
    n = len(y)
    gradient = 0
    for y_i, y_hat_i in zip(list(y),list(y_hat)):
        gradient += np.sign(y_i-y_hat_i)
    return 1/n * gradient
```

```python
k = random.random() * 200 - 100  # -100 100
b = random.random() * 200 - 100  # -100 100

learning_rate = 1e-3

iteration_num = 200 
losses = []
for i in range(iteration_num):
    
    y_hat = f(X)  # \hat{y}
    
    current_loss = loss(y, y_hat)
    losses.append(current_loss)
#     print("Iteration {}, the loss is {}, parameters k is {} and b is {}".format(i,current_loss,k,b))
    
    k_gradient = partial_derivative_k(X, y, y_hat)
    b_gradient = partial_derivative_b(y, y_hat)
    
    k = k + (-1 * k_gradient) * learning_rate
    b = b + (-1 * b_gradient) * learning_rate
best_k = k
best_b = b
```
